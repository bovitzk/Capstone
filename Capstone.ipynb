My (imaginary) child is trying to decide what school to attend. We are accessing the data from the College Scorecard from ED.gov for the 2015-16 year to help her narrow down the field of schools she wants to visit. The set includes data on schools across the country to help students and families compare college costs and outcomes as they weigh the tradeoffs of different colleges, accounting for their own needs and educational goals. These data are provided through federal reporting from institutions, data on federal financial aid, and tax information. These data provide insights into the performance of institutions that receive federal financial aid dollars, and the outcomes of the students of those institutions. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("MERGED2015_16_PP.csv")
```

My (imaginary) daughter is certain she wants to attend school in either Oregon, California or Washington, so we are only going to evaluate schools in those three states.

```python
or_schools = df.loc[lambda d_f: df['STABBR'] == 'OR' , :]
ca_schools = df.loc[lambda d_f: df['STABBR'] == 'CA' , :]
wa_schools = df.loc[lambda d_f: df['STABBR'] == 'WA' , :]
waca_schools = wa_schools.append(ca_schools)
wcs = waca_schools.append(or_schools)
print(wcs["UNITID"].count())
```

Before determining the final scoring matrix with which to rank the schools, we take a look at diversity in the schools. We want to make sure there is not a tradeoff between diversity and academic standards. If there is, we will have to create an appropriate weight for the importance of diversity in our scoring. To look at this we compare the midpoint of the ACT cumulative score for admission to the total share of enrollment of undergraduate degree-seeking students who are white non-Hispanic.

Unfortunately the schools we are looking at do not report a "white non-Hispanic" category, so we have to use white as a proxy category for lack of diversity. 

```python
print(or_schools['UGDS_WHITE'].describe(), wa_schools['UGDS_WHITE'].describe(), ca_schools['UGDS_WHITE'].describe())
```

Unsurprisingly, we see that Oregon schools are the least diverse, on average, and that California has the highest levels of diversity. We will have to include diversity into the scoring matrix to emphasize our interest in finding schools that are diverse.

We need a scoring matrix that will allow a college to score high for having a lower cost, a higher quality of education as determined by entrance scores and completion rates, and more diversity. For cost a school whose cost of tuition is less than half the average will have a score of 0, schools whose cost is more than twice the average will score 2 and schools in between will score 1. 

Quality will be a composite of completion rates and average ACT scores. A college will score 0 if it has the highest test score, and all other scores will be 1 minus the decimal representation of the percentage of the maximum score, so a midpoint score of 35 on the ACT is a score of 0 and a midpoint of 34/35 on the ACT = 1-34/35 or approximately 0.02857142857. The same will happen for each SAT subject test. A 100% completion rate will score 0 points, and all other schools will score 1 minus the completion rate. 

There will be two potential points for diversity. The first will be 1 minus the mean percentage of white undergraduate degree-seeking students, and the other will be 1 minus the median, or 50% percentile.
